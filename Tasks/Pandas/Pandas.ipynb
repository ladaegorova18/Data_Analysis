### Merge таблиц

```python
df1 = pd.DataFrame({
    'car': ['Skoda','Volvo','Lada', 'Renault'],
    'model': ['Octavia','S90','Largus','Sandero'],
    'hp': [150, 235, 106, 106],
    '100km': [8.8, 6.3, 11, 10.3]
    
    
})

df2 = pd.DataFrame({
    'car': ['Skoda','Volvo','Lada', 'Renault'],
    'model': ['Fabia','S60','Granta','Duster'],
    'hp': [125, 190, 82, 82],
    '100km': [10.1, 8.3, 12, 12.3]
    
})


df1.index = [0,1,2,3]
df2.index = [2,3,4,5]
print(df1)
print()
print(df2)
```

```python
pd.concat([df1,df2], axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df2.reset_index()
```

```python
pd.concat([df1.reset_index(),df2.reset_index()],axis=1)
```
